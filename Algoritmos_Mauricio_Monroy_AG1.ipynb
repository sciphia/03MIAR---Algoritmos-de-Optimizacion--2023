```python
#Tower of Hanoi
#The game always have 3 poles and goes from the first to the third pole

def Hanoi_solver(n, a=1, b=3):
    # n is the number of disks
    # a is the starting pole (integer)
    # b is the final destination pole (integer)
    if n == 1:
      print(f"Move disk from {a} to {b}.")
    else:
      Hanoi_solver(n-1,a,6-b-a)
      print(f"Move disk from {a} to {b}.")
      Hanoi_solver(n-1,6-b-a,b)

Hanoi_solver(3)
```

```python
#Coin Change - TÃ©cnica Voraz
currency = [10, 5, 2, 1]

def coin_change(n,currency):
  res = []
  for i,j in enumerate(currency):
    coins = n//j
    res.append(coins)
    n -= coins*j
  print(res)

coin_change(34,currency)
```

```python
#Backtracking Technique

def solutionCheck(step,opt):
  for i in range(step+1):
    if opt.count(opt[i]) > 1:
        return False
  for j in range(i+1, step+1):
    if abs(i-j) == abs(opt[i]-opt[j]):
      return False
  return True

def QueenSolver(n,step=0,opt=[]):
  if len(opt) == 0:
    opt = [0 for i in range(n)]
  for i in range(1,n+1):
    opt[step] = i
    if solutionCheck(step,opt):
      if step == n-1:
        print(opt)
      else:
        QueenSolver(n,step+1,opt)
    else:
      None
  opt[step] = 0


QueenSolver(8)
```
